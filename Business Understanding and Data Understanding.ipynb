```python
def get_mean(dataset, column_name):
    """
    Description: This function can be used to get the mean of a column. 

    Arguments:
        dataset: Dataset 
        column_name: Column name that you want to get the mean.

    Returns:
        The mean of a column located in the dataset
    """
    mean = dataset[column_name].mean()
    return mean

def list_count_variables(dataset, column_name):
    """
    Description: Get a list of a categorical variables and the amount of when each categoriacal variable appears.  

    Arguments:
        dataset: the cursor object. 
        column_name: Column name that you want to get the count.

    Returns:
        The value value_counts of the column
    """
    return dataset[column_name].value_counts()
```
